```python
df = pd.read_sql("SELECT * FROM customers", db, index_col="CustomerId")
df.head()
```

```python
df=pd.read_sql("SELECT CustomerId, FirstName, LastName, Phone FROM customers", db, index_col="CustomerId")
df.head()
```

```python
kamanda="SELECT CustomerId, FirstName, LastName, Phone, Country FROM customers ORDER BY Country"
df=pd.read_sql(kamanda, db, index_col="CustomerId")
df
```
