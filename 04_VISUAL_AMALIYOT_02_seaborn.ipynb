### 3-VAZIFA. Grafiklar nomini to'g'rilang (`T`, `Tm` va hokazo matnlarni o'zbek tilida chiqaring.)

```python
df.columns
```

```python
df=df.rename(columns={'T':'Yillik ortacha harorat', 'TM':'Yillik ortacha maksimal harorat', 'Tm':'Yillik ortacha minimal harorat', 'V':'Yillik ortacha shamol tezligi', 'RA':'Yomgirli kunlar soni', 'SN':'Boronli kunlar soni', 'TS':'Tumanli kunlar soni', 'FG':'Tufonli kunlar soni', 'TN':'Dol yoqqan kunlar'})
plt.figure(figsize=(18,6))
plt.title("Toshkent shahri yillik ob-havo ko'rsatkichlari")
plt.xlabel("Yil")
plt.ylabel("Harorat")
sns.lineplot(data=df)
plt.show()
```

```python
#@title Kutilgan natija:
labels=["Yillik o'rtacha harorat",
        "Yillik o'rtacha maksimal harorat",
        "Yillik o'rtacha minimal harorat",        
        "Yillik o'rtacha shamol tezligi",
        "Yomg'irli kunlar soni",
        "Bo'ronli kunlar soni",
        "Tumanli kunlar soni",
        "To'fonli kunlar soni",
        "Do'l yoqqan kunlar"
        ]

plt.figure(figsize=(16,6))
sns.lineplot(data=df, linewidth=2.0)
plt.title("Toshkent shahri uchun ob havo ko'rsatkichlari")
plt.xlabel("Yil")
plt.ylabel("Harorat")
plt.legend(labels, loc=1)
plt.show()
```

### 4-VAZIFA. Yuqoridagi grafikda turli ko'rsatkichlar aralashib ketdi. Keling ustun turiga qarab alohida grafiklar chizamiz.

```python
plt.figure(figsize=(16,6))
plt.title("Toshkent shahri yillik ob-havo ko'rsatkichlari")
plt.xlabel("Yil")
plt.ylabel("Harorat")
sns.lineplot(data=df[['Yillik ortacha harorat', 'Yillik ortacha maksimal harorat', 'Yillik ortacha minimal harorat']])
plt.show()
```

```python
#@title Kutilgan natija:
labels=["Yillik o'rtacha harorat",
        "Yillik o'rtacha maksimal harorat",
        "Yillik o'rtacha minimal harorat",        
        # "Yillik o'rtacha shamol tezligi",
        # "Yomg'irli kunlar soni",
        # "Bo'ronli kunlar soni",
        # "Tumanli kunlar soni",
        # "To'fonli kunlar soni",
        # "Do'l yoqqan kunlar"
        ]

plt.figure(figsize=(16,6))
sns.lineplot(data=df[['T','TM','Tm']], linewidth=2.0)
plt.title("Toshkent shahri uchun ob havo ko'rsatkichlari")
plt.xlabel("Yil")
plt.ylabel("Harorat")
plt.legend(labels, loc=1)
plt.show()
```

#### 4.2. Kunlarni ko'rsatuvchi ustunlarni alohida grafik ko'rinishida chiqaring

```python
plt.figure(figsize=(16,6))
plt.title("Toshkent shahri yillik ob-havo ko'rsatkichlari")
plt.xlabel("Yil")
plt.ylabel("Kunlar")
sns.lineplot(data=df[['Yomgirli kunlar soni', 'Boronli kunlar soni', 'Tumanli kunlar soni', 'Tufonli kunlar soni','Dol yoqqan kunlar']])
plt.show()
```

### 5-VAZIFA. Yillik o'rtacha shamol tezligini ustunli grafik ko'rinishida chiqaring.

```python
plt.figure(figsize=(16,6))
plt.title("Toshkent shahri yillik ob-havo ko'rsatkichlari")
plt.xlabel("Yil")
plt.ylabel("O'rtacha shamol tezligi (km/soat)")
sns.barplot(x=df.index, y=df['Yillik ortacha shamol tezligi'])
plt.show()

```

```python
#@title Kutilgan natija:
plt.figure(figsize=(16,6))
sns.barplot(x=df.index, y=df['V'])
plt.title("Toshkent shahri uchun ob havo ko'rsatkichlari")
plt.ylabel("O'rtacha shamol tezligi (km/soat)")
plt.xlabel("Yillar")
plt.xticks(rotation=90)
plt.show()
```

### 6-VAZIFA. Yomg'rili, qorli, bo'ronli, tumanli va do'l yoqqan kunlar sonini heatmap ko'rinishida chiqaring.

```python
df=df.rename(columns={'Yomgirli kunlar soni':'Yomgir','Boronli kunlar soni':'Buron', 'Tumanli kunlar soni':'Tuman', 'Tufonli kunlar soni':'Qor', 'Dol yoqqan kunlar':'Dol'})
plt.figure(figsize=(30,6))
plt.title("Toshkent shahri yillik ob-havo ko'rsatkichlari")

sns.heatmap(data=df[['Yomgir','Buron', 'Tuman', 'Qor', 'Dol' ]].T, label=1, cmap="Accent", linewidth=1, annot=True)
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.show()

```

```python
#@title Kutilgan natija:
plt.figure(figsize=(30,5))
sns.heatmap(data=df[['RA','SN','TS','FG','GR']].T, cmap="YlGnBu", linewidths=1, square=True)
plt.title("Toshkent shahri uchun ob havo ko'rsatkichlari")
yticks_labels = ["Yomg'ir","Qor","Bo'ron","Tuman","Do'l"]
plt.yticks(np.arange(5)+0.5, labels=yticks_labels)
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.show()
```
